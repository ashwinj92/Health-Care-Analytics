```python
path = "/home/ronaldsonbellande/Desktop/Healthcare Data Analytics/HealthCare Analysis Group/Health-Care-Analytics/Activity Data/"
data_training = pd.read_csv(path + "train.csv")
data_test = pd.read_csv(path + "test.csv")
```

```python
Labels = data_training['activity']
data_training_label = data_training.drop(['rn', 'activity'], axis = 1)
Labels_keys = Labels.unique().tolist()
Labels = np.array(Labels)
print('Activity labels: ' + str(Labels_keys))
```

```python
X_axis = data_training.drop("activity", axis=1)
X_axis
```

```python
Y_axis = data_training["activity"].copy()
Y_axis
```

```python
Y_axis = pd.get_dummies(Y_axis)
Y_axis
```

```python
Y_axis = np.argmax(Y_axis.values, axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_axis.values, Y_axis, random_state=0)
```

```python
y_train
```

```python
X_train_data= data_training.iloc[:,:-2]
Y_train_data=data_training.iloc[:,1]
X_test_data= data_test.iloc[:,:-2]
Y_test_data=data_test.iloc[:,-1]
```

```python
Y_train_data
```

```python
Y_training_data_count=np.array(Y_train_data.value_counts())
activity=sorted(Y_train_data.unique())
```

```python
Y_training_data_count
```

```python
plt.figure(figsize=(15,5))
plt.pie(Y_training_data_count,labels=activity)
plt.savefig("/home/ronaldsonbellande/Desktop/Healthcare Data Analytics/HealthCare Analysis Group/Health-Care-Analytics/Models and Data/", dpi=500)

plt.show()
```

```python
model = Sequential()
model.add(Dense(32,activation='sigmoid',input_dim=562))
model.add(Dropout(0.5))
model.add(Dense(6,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
y_trained = keras.utils.to_categorical(y_train, num_classes=6)
```

```python
activity_model_recognition = model.fit(X_train, y_trained, epochs=2000,
                                       batch_size=64, verbose=2)
```

```python
path_for_saving = "/home/ronaldsonbellande/Desktop/Healthcare Data Analytics/HealthCare Analysis Group/Health-Care-Analytics/Models and Data/"
```

```python
model.save(path_for_saving + 'activity_recognition_model.h5')
```

```python
y_test_trained = keras.utils.to_categorical(y_test, num_classes=6)
model.evaluate(X_test,y_test_trained)
```

```python
prediction = model.predict(X_test)
confusion = confusion_matrix(y_test, np.argmax(prediction,axis=1))
print(confusion_matrix(y_test, np.argmax(prediction,axis=1)))
print(confusion_matrix(y_test, np.argmax(prediction,axis=1)))
print(classification_report(y_test,  np.argmax(prediction,axis=1), target_names=["LAYING", "SITTNG", "STANDNG", "WALKING", "WALKING_DOWNSTAIRS", "WALKING_UPSTAIRS"]))
```

```python
rcParams['figure.figsize'] = 10, 4
plt.plot(activity_model_recognition.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.savefig(path_for_saving + 'model_accuracy.png', dpi=500)
plt.show()
# summarize history for loss
plt.plot(activity_model_recognition.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.savefig(path_for_saving + 'model_lost.png', dpi=500)
plt.show()
```
